```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

# plotting defaults
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (7, 5)

outcome_var = "LBW_Rate"

id_vars = ["ZIP_Code_of_Residence"]

population_count_vars = ["Total_Births_2018_2022","Total_LBW_Count_2018_2022"]

predictor_vars = [
    # Environmental
    "avg_traffic_pct","avg_diesel_pm","avg_cancer_risk","avg_resp_hazard","avg_ej_index","combined_noise_mean_db",

    # SVI / indices
    "svi_score","socioeconomic_status","household_characterisitcs","race_ethnicity_minority_status","housing_type_transportation",

    # SES / housing
    "educational_attainment","homeownership_rate","poverty_rate","log_median_household_income",

    # Demographics
    "white_prop","black_prop","asian_prop","hispanic_prop","insurance_coverage_prop"
]

# Leakage & consistency checks

# Outcome must not be a predictor
assert outcome_var not in predictor_vars

# ID variables must not be predictors
assert not set(id_vars).intersection(predictor_vars)
```

```python
data_path = r"C:\Users\Elias\Final Project\Cleaned output data files\master_data_with_noise_2022.csv"
df = pd.read_csv(data_path)
df.shape
df.head()
df.info()
```

```python
numeric_cols = df.columns.drop(df.select_dtypes(include=["object", "category"]).columns)

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
percent_cols = [
    c for c in df.columns
    if ("percent" in c.lower()) or ("rate" in c.lower())
]

percent_cols = [c for c in percent_cols if c not in [outcome_var]]

for col in percent_cols:
    df.loc[(df[col] < 0) | (df[col] > 100), col] = np.nan
```

```python
# checking for missing data

missing_summary = (
    df.isna()
      .mean()
      .reset_index()
      .rename(columns={"index": "feature", 0: "missing_rate"})
      .sort_values("missing_rate", ascending=False)
)

missing_summary
```

```python
# classifying missing data
keep_low_missing = missing_summary.loc[missing_summary["missing_rate"] < 0.05, "feature"].tolist()

review_mid_missing = missing_summary.loc[(missing_summary["missing_rate"] >= 0.05) & (missing_summary["missing_rate"] <= 0.20), "feature"].tolist()

candidate_drop_high_missing = missing_summary.loc[missing_summary["missing_rate"] > 0.20, "feature"].tolist()
```

```python
# Univariate EDA
univariate_summary = (df[predictor_vars + [outcome_var]].describe().T)

univariate_summary["IQR"] = (univariate_summary["75%"] - univariate_summary["25%"])

univariate_summary

# creating distribution plots

sns.set(style="whitegrid")

for col in predictor_vars + [outcome_var]:
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    sns.histplot(df[col], kde=True, ax=axes[0])
    axes[0].set_title(f"Histogram: {col}")

    sns.boxplot(x=df[col], ax=axes[1])
    axes[1].set_title(f"Boxplot: {col}")

    plt.tight_layout()
    # saving to current directory
    plt.savefig(f"univariate_{col}.png", dpi=300, bbox_inches='tight')
    plt.show()
    plt.close()
```

```python
# Coorelation analysis
# Which predictors are associated with LBW_rate, and how strong are those associations?
# Using Spearman coorelation (rank-based) because many predictors are skewed + LBW_Rate is bounded from 0 to 100

from scipy.stats import pearsonr, spearmanr

corr_results = []

for col in predictor_vars:
    pearson_r, pearson_p = pearsonr(df[col], df[outcome_var])
    spearman_r, spearman_p = spearmanr(df[col], df[outcome_var])

    corr_results.append({
        "feature": col,
        "pearson_r": pearson_r,
        "pearson_p": pearson_p,
        "spearman_r": spearman_r,
        "spearman_p": spearman_p
    })

corr_results = pd.DataFrame(corr_results)
corr_results.sort_values("pearson_p")
```

```python
import statsmodels.api as sm

univariate_reg_results = []

for col in predictor_vars:
    X = sm.add_constant(df[col])
    y = df[outcome_var]

    model = sm.OLS(y, X).fit()

    univariate_reg_results.append({
        "feature": col,
        "coef": model.params[col],
        "p_value": model.pvalues[col],
        "r_squared": model.rsquared
    })

univariate_reg_results = pd.DataFrame(univariate_reg_results)
univariate_reg_results.sort_values("p_value")
```

```python
# full regression to determine multicollinearity
X_full = sm.add_constant(df[predictor_vars])
y = df[outcome_var]

full_model = sm.OLS(y, X_full).fit()
print(full_model.summary())
```

```python
# Predictor correlation heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(
    df[predictor_vars].corr(),
    cmap="coolwarm",
    center=0,
    linewidths=0.5,
    annot=True,       
    fmt=".2f",        
    annot_kws={"size":10}  
)
plt.title("Predictor Correlation Heatmap")
plt.tight_layout()
plt.show()

# Correlation matrix as a table
corr_table = df[predictor_vars].corr(method="spearman")

corr_table
```

```python
# Variance Inflation Factor
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_vif = sm.add_constant(df[predictor_vars])

vif_table = pd.DataFrame({
    "feature": X_vif.columns,
    "VIF": [
        variance_inflation_factor(X_vif.values, i)
        for i in range(X_vif.shape[1])
    ]
})

vif_table = vif_table[vif_table["feature"] != "const"]
vif_table.sort_values("VIF", ascending=False)
```

```python
decision_table = (
    corr_results
    .merge(univariate_reg_results, on="feature")
    .merge(vif_table, on="feature")
)

decision_table["cluster"] = ""
decision_table["decision"] = ""

decision_table
```

```python
# Decision table

decision_table = (corr_results.merge(univariate_reg_results, on="feature").merge(vif_table, on="feature"))
decision_table["cluster"] = ""
decision_table["decision"] = ""

# Assigning clusters based on coorelation greater than or equal to 0.7 (Heatmap)

cluster_map = {
    # cluster_air_pollution
    "avg_traffic_pct": "cluster_air_pollution",
    "avg_diesel_pm": "cluster_air_pollution",
    "avg_cancer_risk": "cluster_air_pollution",
    "avg_resp_hazard": "cluster_air_pollution",

    # cluster_social_ej
    "avg_ej_index": "cluster_social_ej",
    "svi_score": "cluster_social_ej",
    "socioeconomic_status": "cluster_social_ej",
    "household_characteristics": "cluster_social_ej",
    "race_ethnicity_minority_status": "cluster_social_ej",
    "housing_type_transportation": "cluster_social_ej",
    "educational_attainment": "cluster_social_ej",
    "log_median_household_income": "cluster_social_ej",
    "poverty_rate": "cluster_social_ej",
    "white_prop": "cluster_social_ej",
    "hispanic_prop": "cluster_social_ej",
}

decision_table["cluster"] = decision_table["feature"].map(cluster_map).fillna("solo")

# Applying decision rules

#Rule 1: drop for no outcome signal if pearson and spearman p-values are greater than and equal to 0.10
rule1_mask = (
    (decision_table["pearson_p"] > 0.10) &
    (decision_table["spearman_p"] > 0.10)
)

decision_table.loc[rule1_mask, "decision"] = "Drop"


# Handling clustered features
for cluster_name in decision_table["cluster"].unique():

    if cluster_name == "solo":
        continue

    cluster_df = decision_table.query("cluster == @cluster_name")

    # Exclude already dropped by Rule 1
    cluster_df = cluster_df[cluster_df["decision"] == ""]

    if cluster_df.empty:
        continue

    # Strongest univariate signal = smallest univariate_p
    keep_feature = cluster_df.sort_values("p_value").iloc[0]["feature"]

    decision_table.loc[decision_table["feature"] == keep_feature, "decision"] = "Keep"

    redundant_features = cluster_df[cluster_df["feature"] != keep_feature]["feature"]

    for feat in redundant_features:
        vif_value = decision_table.loc[decision_table["feature"] == feat, "VIF"].iloc[0]

        if vif_value >= 10:
            decision_table.loc[
                decision_table["feature"] == feat, "decision"] = "Merge_pca"
        else:
            decision_table.loc[
                decision_table["feature"] == feat, "decision"] = "Drop"


# Solo features
solo_mask = (
    (decision_table["cluster"] == "solo") &
    (decision_table["decision"] == "")
)

decision_table.loc[solo_mask, "decision"] = "Keep"




decision_table
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

features_marked_Drop = (decision_table.query("decision == 'Drop'")["feature"].tolist())

pca_features = (decision_table.query("decision == 'Merge_pca'")[["feature", "cluster"]])

for cluster_name in pca_features["cluster"].unique():

    cluster_feats = (pca_features.query("cluster == @cluster_name")["feature"].tolist())

    # Standardize
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(df[cluster_feats])

    # PCA
    pca = PCA(n_components=1)
    pc1 = pca.fit_transform(X_scaled)

    # Add PC1 to dataframe
    df[f"pca_{cluster_name}"] = pc1.flatten()

    # Drop original features
    df = df.drop(columns=cluster_feats)
```

```python
final_predictors = (decision_table.query("decision == 'Keep'")["feature"].tolist())

final_pca_features = [col for col in df.columns if col.startswith("pca_")]

final_features = final_predictors + final_pca_features
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
X = df[final_features]
X = sm.add_constant(X)

y = df["LBW_Rate"]

model = sm.OLS(y, X).fit()
print(model.summary())


vif_df = pd.DataFrame()
vif_df["feature"] = X.columns
vif_df["VIF"] = [
    variance_inflation_factor(X.values, i)
    for i in range(X.shape[1])
]

vif_df
```

```python
# Final columns
id_vars = ["ZIP_Code_of_Residence"]
outcome_var = "LBW_Rate"

final_predictors = ["combined_noise_mean_db","homeownership_rate","log_median_household_income","black_prop","asian_prop","insurance_coverage_prop","pca_cluster_social_ej"]

# final dataset
final_df = df[id_vars +["Total_Births_2018_2022"] +   # kept for reference only
    [outcome_var] +final_predictors].copy()

final_df.to_csv("final_model_dataset.csv", index=False)
```
