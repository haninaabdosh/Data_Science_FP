```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir("C:/Users/Demoted/Downloads/datasci/Data_Science_FP")

df = pd.read_csv("Cleaned output data files/final_model_dataset.csv")
```

```python
plt.figure(figsize=(12, 7))

x = df["combined_noise_mean_db"]
y = df["LBW_Rate"]

plt.scatter(x, y, alpha=0.6, s=80, color="#3b82f6", edgecolors="black", linewidth=0.5)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", linewidth=2, label=f'Trend line: y={z[0]:.4f}x+{z[1]:.2f}')

correlation = df['combined_noise_mean_db'].corr(df['LBW_Rate'])

plt.xlabel('Noise Level (dB)', fontsize=12, fontweight='bold')
plt.ylabel('Low Birth Weight Rate (%)', fontsize=12, fontweight='bold')
plt.title('Relationship: Prenatal Noise Exposure vs Low Birth Weight Rate\n(LA County, 2018-2022)',
fontsize=14, fontweight='bold', pad=20)
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=10)

textstr = f'Pearson r = {correlation:.3f}\nn = {len(df)} ZIP codes'
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, fontsize=10,
verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()

os.makedirs('visualizations/visualization_media', exist_ok=True)
plt.savefig('visualizations/visualization_media/viz_01_noise_vs_lbw_scatter.png', dpi=300, bbox_inches='tight')

print("âœ“ Visualization 1 saved: visualizations/visualization_media/viz_01_noise_vs_lbw_scatter.png")

plt.show()
```
